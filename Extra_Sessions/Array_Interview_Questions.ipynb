```python
# 1. Find the kth largest/smallest item from a list

L = [12,23,1,4,56,34,22,3]

L.sort()
k = int(input("Enter the value of k : "))
print(f"The {k}th smallest value of the List {L} is {L[k-1]}")
```

```python
# 2. Check if an array is sorted
L = [1,2,3,7,5]

flag = True
for i in range(len(L) - 1):
    if L[i] > L[i+1]:
        flag = False
        break
    
if flag:
    print("Sorted")
else:
    print("Not Sorted")
```

```python
# 3. Find Min/Max in a given array

L = [21,1,34,23,54,11,10]

max_value = L[0]

for number in L:
    if number > max_value:
        max_value = number

print(max_value)
```

```python
# 4. Find the first element to occur k times in an array

L = [ 1, 1, 2, 3, 4, 4, 4, 5, 5, 6, 6, 6]

k = 3

res = {}

for number in L:
    if number in res:
        res[number] += 1
    else:
        res[number] = 1

for key in res:
    if res[key] == k:
        print(key)
        break
```

```python
# 5. Find duplicates in an array
L = [1,1,2,3,4,4,5,5]

res = {}

for number in L:
    if number in res:
        res[number] += 1
    else:
        res[number] = 1

for key in res:
    if res[key] > 1:
        print(key)
```

```python
# 6. Rotate array to left d items

L = [1,2,3,4,5]
rotate = 2

for i in range(rotate):
    temp = L[0]
    for j in range(0, len(L)-1):
        L[j] = L[j+1]
    L[-1] = temp
    

print(L)
```

```python
# 8. Find continous subarray with a given sum(given non-negative numbers)
# return the starting and ending index of the subarray
# return 1st subarray in case of multiple

L = [-2,4,7,-1,6,-11,14,3,-1,-6]
S = 22
res = {}

for i in range(0, len(L)):
    subarray = []
    for j in range(i, len(L)):
        subarray.append(L[j])
        res[tuple(subarray)] = sum(subarray)

for key in res:
    if res[key] == S:
        print(key)
```

```python
# 9. Find element with left side smaller/right side greater in an array

L = [3,1,2,5,8,7,9]

for i in range(1,len(L)-1):
    if max(L[:i]) < L[i] < max(L[i+1:]):
        print(L[i])
```

```python
# 10. Maximum sum subarray
L = [-2,4,7,-1,6,-11,14,3,-1,-6]
S = 22
res = {}
for i in range(0, len(L)):
    subarray = []
    for j in range(i, len(L)):
        subarray.append(L[j])
        res[tuple(subarray)] = sum(subarray)
        
list(sorted(res.items(),key = lambda x:x[1], reverse= True)[0][0])
```

```python
# Kadance Algorithm

L = [-2,4,7,-1,6,-11,14,3,-1,-6]

curr_sum = 0
best_sum = float('-inf')
curr_seq = []
best_seq = []

for number in L:
    if number + curr_sum < number:
        curr_sum = number
        curr_seq = [number]
    else:
        curr_sum = curr_sum + number
        curr_seq.append(number)
    
    if best_sum < curr_sum:
        best_sum = curr_sum
        best_seq = curr_seq.copy()

print(best_seq, best_sum)
```
