### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
x1 = int(input("Enter the x-co-ordinate of Point 1"))
y1 = int(input("Enter the y-co-ordinate of Point 2"))
x2 = int(input("Enter the x-co-ordinate of Point 1"))
y2 = int(input("Enter the y-co-ordinate of Point 2"))

euclidean_distance = ( ( x1 - x2 ) ** 2 + (y1 - y2) ** 2 ) ** 0.5 

print(euclidean_distance)
````

### Q5:- Write a program to find the simple interest when the value of principle,rate of interest and time period is provided by the user.


````python
p = int(input("Enter the principal"))
r = int(input("Enter the rate of interest"))
t = int(input("Enter the time"))

simple_interest = ( p * t * r)/100
print(simple_interest)
````

### Q8:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
n1 = int(input("Enter the numerator of the 1st fraction"))
d1 = int(input("Enter the denominator of the 1st fraction"))
n2 = int(input("Enter the numerator of the 2nd fraction"))
d2 = int(input("Enter the denominator of the 2nd fraction"))

den = d1 * d2
num = ( n1 * d2 ) + ( n2 * d1 )

result = num / den
print(result)
````
